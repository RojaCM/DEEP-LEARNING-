```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')

query_embedding = model.encode('Istio Service Mesh')
passage_embedding = model.encode(["Tech Live Session | Istio Service Mesh",'London is known for its finacial district'])

print("Similarity:", util.dot_score(query_embedding, passage_embedding))
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')

query_embedding = model.encode('Altimetrik in  fashion and apparel industry')
passage_embedding = model.encode(["The fashion and apparel industry is undeniably one of the most dynamic businesses to be in. It is in constant race with itself to keep updating styles, trends, patterns, colors and products. What's in style in summer, has completely changed by fall. What makes it even more difficult to keep up is public opinion. Regardless of what fashion designers decide for the global market, the ultimate winner is what the customer chooses to buy and wear. Even if cerulean gets to be the color of the season, customers may want their preferred products in their favorite shade of blue or yellow. As a leader in this industry, the client arrived at the pathbreaking idea of co-creating products, i.e., evolve products in tandem with actual customer feedback. A technology platform that enables this idea would eventually also lead to optimizing merchandizing and reducing inventory."])

print("Similarity:", util.dot_score(query_embedding, passage_embedding))
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')

query_embedding = model.encode("why are rolls royce so expensive???")
passage_embedding = model.encode(["Rolls-Royce makes some of the world's most luxurious cars. Known for producing hand-crafted automobiles that promise a seamless \"magical carpet ride\" for its customers, a Rolls-Royce car does not come cheap. These are some of the best-selling Rolls-Royce models, and these are their entry-level ....."])

print("Similarity:", util.dot_score(query_embedding, passage_embedding))
```
