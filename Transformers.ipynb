```python
!pip install transformers
from transformers import pipeline
```

```python
nlp=pipeline("question-answering")
```

```python
context=r"""Altimetrik India Private Limited is an unlisted private company incorporated on 21 July, 2003. It is classified as a private limited company and is located in Bangalore, Karnataka. It's authorized share capital is INR 29.50 cr and the total paid-up capital is INR 27.52 cr."""
```

```python
nlp(question="where is Altimertik office located",context=context)
```

```python
nlp(question="where is Altimetrik located ",context=context)
```

```python
nlp1=pipeline("ner", grouped_entities=True)
```

```python
sequence ="Altimetrik has come up with a solution to prevent DDoS attack"
```

```python
output=nlp1(sequence)
```
