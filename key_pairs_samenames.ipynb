```python
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
```

```python
pip install keybert
```

```python
kw_model=KeyBERT(model)
```

```python
keywords_score=kw_model.extract_keywords("The Future Trends In Mobility And Transportation",keyphrase_ngram_range=(1,3))
keywords=[]
scores=[]
for i ,j in keywords_score:
  keywords.append(i)
  scores.append(j)


extracted_keywords=[]
for keyword,boost in zip(keywords,scores):
  extracted_keywords.append(
      {"keyword":keyword,
       "boost":boost})

extracted_keywords
```
